This code file is to expermiment with how to make a live to-do lis for the discord bot

Features: 

-- Update to-do list with new entry

-- Send regular reminders based on user preference

-- Adjust priority (easily)

-- Mark different stages and track completion

Things to add to To-Do List:


CODING PROJECTS:
- Code Learning Tracker
- Peronsal Health Tracker (workouts, sleep, etc)
- Financial Tracker
- General Video Game Tracker (100%, game-picker)





```python
# Importing Libaries


import numpy as np
import pandas as pd

import datetime
import os

```

```python
to_list_pd = pd.DataFrame(
    {
        "TASK": "string",
        "TASK CREATION": pd.to_datetime(datetime.datetime.now().isoformat(' ', 'seconds')),

        "CATAGORY": pd.Categorical(["None"]),
        "GROUP": "None",
        "SUB-GROUP": "None",

        "RELEVANT LINK": 'https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.to_markdown.html', 

        "RECURRING": False,
        "RECURRING INTERVAL": [24],
        "DUE DATE": pd.to_datetime("20130102"),

        "PRIORITY": [10],
        "STATUS": pd.Categorical(["Not Started"], 
                                 categories = [
                                     "Not Started",
                                     "In Progress",
                                     "Pending",
                                     "Blocked",
                                     "Hiatus",
                                     "Completed"
                                               ], 
                                 ordered= True),
        "START TIME": pd.Timestamp("20130102"),
        "ESTIMATED TIME": [24],
        "LOGGED HOURS": [24],

        "COMPLETED": True,
        "COMPLETED TIME": pd.Timestamp("20130102")


    }
)

# pd.Categorical(["test", "train", "test", "train"]),
to_list_pd
```
